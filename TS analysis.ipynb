# Simple Moving Average

```python
df_dt["SMA-Open"] = df_dt["Open"].rolling(window=50,min_periods=1).mean()

df_dt
```

```python
df_dt[["Open","SMA-Open"]].plot(xlim=("2019-01-01","2022-01-01"))
```

# Cummulative Moving Average

```python
df_dt["CMA-High"] = df_dt["High"].expanding().mean()

df_dt
```

```python
df_dt[["High","CMA-High"]].plot(xlim=("2019-01-01","2022-01-01"))
```

```python
import pandas as pd

# Sample DataFrame with a date string
df = pd.DataFrame({
    'date_string': ['2023-08-21 14:30:00', '2023-08-22 15:45:00']
})

# Convert the string to datetime
df['date'] = pd.to_datetime(df['date_string'])

# Convert the datetime back to string with desired format
df['date_formatted'] = df['date'].dt.strftime('%Y-%m-%d %H:%M')

print(df)
```
